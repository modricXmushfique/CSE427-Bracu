# **Numpy Task**

```python
#1 Create two linear arrays a and b with 20 random elements each ranging from 1 to 100.

import numpy as np

a = np.random.randint(1, 101, 20)    #a,b duita linear array create with 20 random element range(1-100)
b = np.random.randint(1, 101, 20)

```

```python
#2 Now change those linear arrays to two 5x4 matrices.

a = a.reshape(5, 4)          #reshape the linear array into 5x4 matrices
b = b.reshape(5, 4)

print("Matrix a:")         #print function
print(a)
print("\nMatrix b:")
print(b)
```

```python
#3 Perform scalar multiplication with matrix a and b.


a_scaled = a * 2          # Perform scalar multiplication
b_scaled = b * 3
```

```python
#4 Now try to modify the matrix b and perform matrix multiplication with a and b to store it in c.

b_modified = b + 10     ## Modify matrix b by adding a constant (e.g., 10)
c = a @ b_modified.T

# Print the resulting matrix c
print("\nMatrix a after scalar multiplication by 2:")
print(a_scaled)
print("\nMatrix b after scalar multiplication by 3:")
print(b_scaled)
print("\nMatrix b after adding 10:")
print(b_modified)
print("\nResulting matrix c (a @ b_modified.T):")
print(c)
```

```python
#5 Print only the column 2, 3 of matrix c. Print the whole matrix using slicing technique.


print("\nColumns 2 and 3 of matrix c:")       ## Print columns 2 and 3 (index 1 and 2) of matrix c using slicing.
print(c[:, 1:3])

# Print the whole matrix c using slicing.
print("\nWhole matrix c using slicing:")
print(c[:, :])
```

```python
#6 Find the largest element in the array, and also the index of that element.


largest_element = np.max(c)
largest_element_index = np.argmax(c)

print("\nLargest element in matrix c:", largest_element)
# np.argmax returns the flattened index, convert it to 2D index
largest_element_2d_index = np.unravel_index(largest_element_index, c.shape)
print("Index of the largest element:", largest_element_2d_index)
```

```python
#7 Convert the matrix c into an one dimensional array keeping the original array unchanged.


c_flattened = c.flatten()

print("\nMatrix c (original):")
print(c)
print("\nMatrix c converted to a one-dimensional array:")
print(c_flattened)
```

# **Pandas Task**

```python
#1 CSV file:  https://drive.google.com/file/d/1zo5y462G1FaGUlAEUayACHSV2zSliR-5/view?usp=drive_link
#Read the CSV file as DataFrames.

import pandas as pd

url = "https://drive.google.com/file/d/1zo5y462G1FaGUlAEUayACHSV2zSliR-5/view?usp=drive_link"
# The URL is a Google Drive link, and direct reading might not work.
# We need to modify the URL to access the raw CSV content.
# The pattern for a shareable link is https://drive.google.com/file/d/FILE_ID/view?usp=drive_link
# The pattern for direct download is https://drive.google.com/uc?export=download&id=FILE_ID
file_id = url.split('/')[-2]
download_url = f'https://drive.google.com/uc?export=download&id={file_id}'

df = pd.read_csv(download_url)
```

```python
#2  Show the first few rows of the DataFrame to get an idea of the whole dataset.
#3 Show the summarized details of the dataset.

display(df.head())
display(df.info())
```

```python
#4 Show how many null values we have in each column.
#5 Replace any column’s null values with the mean value of that particular column.

display(df.isnull().sum())

# Replace null values with the mean of the respective column
for col in df.columns:
    if df[col].isnull().any():
        df[col].fillna(df[col].mean(), inplace=True)

display(df.isnull().sum())
```

```python
# Calculate the mean of each numerical column in the DataFrame df.
column_means = df.mean(numeric_only=True)

# Fill any null values in the DataFrame df with the calculated mean of their respective columns.
df.fillna(column_means, inplace=True)

# After filling the null values, display the first few rows of the DataFrame again to verify the changes.
display(df.head())

# Display the count of null values in each column again to confirm that all null values have been successfully replaced.
display(df.isnull().sum())
```

# **Matplotlib Task**

```python
#1 Generate a scatter plot graph for temperature_dhaka vs humidity_dhaka using (*) as the marker in color red.

import matplotlib.pyplot as plt

# Define the lists temperature_dhaka and humidity_dhaka
temperature_dhaka = [25,34,21,45,28,6,43,18,7,2]
humidity_dhaka = [28, 25,29,20, 26, 50, 19, 29, 52, 55]

# Create a scatter plot
plt.scatter(temperature_dhaka, humidity_dhaka, marker='*', c='red')

# Add title and labels
plt.title("Temperature vs. Humidity in Dhaka")
plt.xlabel("Temperature")
plt.ylabel("Humidity")

# Display the plot
plt.show()
```

```python
#2 Now do a line plot for study_hours vs marks with figure size (12,8)


# Define the lists study_hours and marks
study_hours = [2,3,4,4, 5, 6, 7, 7, 8, 9, 9, 10, 11, 11, 12]
marks = [6, 10, 15, 20, 34, 44, 55, 60, 55, 67, 70, 80, 90, 99, 100]

# Create a Matplotlib figure with a specified size
plt.figure(figsize=(12, 8))

# Create a line plot
plt.plot(study_hours, marks)

# Add a title and labels to the plot
plt.title("Study Hours vs. Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")

# Display the plot
plt.show()
```

```python
#3 Now draw a horizontal bar graph showing the scores in each of the subjects.
#4 Draw a vertical bar graph where each subject score is visualized in different colors.

# Define the lists subjects and marks
subjects = ['Maths', 'English', 'Science', 'Physics', 'Computer']
marks = [89, 90, 45, 78, 99]

# Create a horizontal bar graph
plt.figure(figsize=(10, 5)) # Adjust figure size for horizontal bars
plt.barh(subjects, marks, color='skyblue') # Use a single color for horizontal bars
plt.title("Subject Scores (Horizontal Bar Graph)")
plt.xlabel("Marks")
plt.ylabel("Subjects")

# Create a vertical bar graph
plt.figure(figsize=(10, 5)) # Adjust figure size for vertical bars
colors = ['red', 'green', 'blue', 'purple', 'orange'] # List of different colors
plt.bar(subjects, marks, color=colors) # Use different colors for vertical bars
plt.title("Subject Scores (Vertical Bar Graph)")
plt.xlabel("Subjects")
plt.ylabel("Marks")

# Display the plots
plt.show()

```

```python
#5 Show all these graphs in one graph using subplot( ) function.

# Define the lists for scatter plot
temperature_dhaka = [25,34,21,45,28,6,43,18,7,2]
humidity_dhaka = [28, 25,29,20, 26, 50, 19, 29, 52, 55]

# Define the lists for line plot
study_hours = [2,3,4,4, 5, 6, 7, 7, 8, 9, 9, 10, 11, 11, 12]
marks_line = [6, 10, 15, 20, 34, 44, 55, 60, 55, 67, 70, 80, 90, 99, 100]

# Define the lists for bar graphs
subjects = ['Maths', 'English', 'Science', 'Physics', 'Computer']
marks_bar = [89, 90, 45, 78, 99]

# Create a figure and a set of subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Scatter plot
axes[0, 0].scatter(temperature_dhaka, humidity_dhaka, marker='*', c='red')
axes[0, 0].set_title("Temperature vs. Humidity in Dhaka")
axes[0, 0].set_xlabel("Temperature")
axes[0, 0].set_ylabel("Humidity")

# Line plot
axes[0, 1].plot(study_hours, marks_line)
axes[0, 1].set_title("Study Hours vs. Marks")
axes[0, 1].set_xlabel("Study Hours")
axes[0, 1].set_ylabel("Marks")

# Horizontal bar graph
axes[1, 0].barh(subjects, marks_bar, color='skyblue')
axes[1, 0].set_title("Subject Scores (Horizontal Bar Graph)")
axes[1, 0].set_xlabel("Marks")
axes[1, 0].set_ylabel("Subjects")

# Vertical bar graph
colors = ['red', 'green', 'blue', 'purple', 'orange']
axes[1, 1].bar(subjects, marks_bar, color=colors)
axes[1, 1].set_title("Subject Scores (Vertical Bar Graph)")
axes[1, 1].set_xlabel("Subjects")
axes[1, 1].set_ylabel("Marks")

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the combined plot
plt.show()
```
